```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = [10,5]
# Ignore warnings

import warnings
# Set the warning filter to ignore FutureWarning
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
x = df.drop('Price', axis = 1)
y = df['Price']
```

# **Standarizing X features**

Data Pre-processing, **standardization** to make values  in range of 0-1

```python
from sklearn import preprocessing
pre_process = preprocessing.StandardScaler().fit(x)
x_trans = pre_process.fit_transform(x)
x_trans
```

```python
x_trans.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_trans, y, test_size = 0.10, random_state = 101)
```

```python
# Fit
# Import model
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Create instance of model
lin_reg = LinearRegression()

# Pass training data into model
lin_reg.fit(x_train, y_train)
# pipe = make_pipeline(StandardScaler(), LinearRegression())
# pipe.fit(x_train, y_train)
```

```python
# Predict
y_pred = lin_reg.predict(x_test)
print(y_pred.shape)
print(y_pred)
```

```python
sns.scatterplot(x=y_test, y=y_pred, color='blue', label='Actual Data points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Ideal Line')
plt.legend()
plt.show()
```

```python
# Combine actual and predicted values side by side
results = np.column_stack((y_test, y_pred))

# Printing the results
print("Actual Values  |  Predicted Values")
print("-----------------------------")
for actual, predicted in results:
    print(f"{actual:14.2f} |  {predicted:12.2f}")
```

```python
residual = actual- y_pred.reshape(-1)
print(residual)
# Distribution plot for Residual (difference between actual and predicted values)
sns.distplot(residual, kde=True)
```

```python
# Score It
from sklearn.metrics import mean_squared_error

print('Linear Regression Model')
# Results
print('--'*30)
# mean_squared_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print evaluation metrics
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor

dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train, y_train)
```

```python
dt_pred = dt_regressor.predict(x_test)
dtM = mean_squared_error(dt_pred, y_test)
print(dtM)
```

# **Random Forest**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train , y_train)

rf_pred = rf_regressor.predict(x_test)
rfM = mean_squared_error(rf_pred, y_test)
print(rfM)
```

# **Gradient Boosting Regression**

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_regressor =  GradientBoostingRegressor()
gb_regressor.fit(x_train, y_train)

gb_pred = gb_regressor.predict(x_test)
gbM = mean_squared_error(gb_pred, y_test)
print(gbM)
```

```python
# Sample model scores (replace these with your actual model scores)
model_scores = {
    "Linear Regression": lrM ,
    "Descison Tree": dtM,
    "Random Forest":rfM,
    "Gradient Boosting": gbM
}

# Sort the model scores in ascending order based on their values (lower values first)
sorted_scores = sorted(model_scores.items(), key=lambda x: x[1])

# Display the ranking of the models
print("Model Rankings with both X & Y standardized (lower values are better):")
for rank, (model_name, score) in enumerate(sorted_scores, start=1):
    print(f"{rank}. {model_name}: {score}")
```
